# Deliverable 1: Retrieve Weather Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

### Generate a set of 2,000 random latitudes and longitudes

```python
# We'll add the code that generates the latitudes and longitudes
#we'll pack the latitudes (lats) and longitudes (lngs) as pairs by zipping them
#(lat_lngs) with the zip() function.

lats = np.random.uniform(low = -90.000, high = 90.000, size = 2000)
lngs = np.random.uniform(low = -90.000, high = 90.000, size = 2000)
lats_lngs = zip(lats, lngs)
lats_lngs
```

```python
# Add the latitudes and longitudes to a list to be able to use them.
coordinates = list(lats_lngs)
```

```python
len(coordinates)
```

### Retrieve the nearest city

```python
# Use the citipy module to determine city based on latitude and longitude.
from citipy import citipy
```

```python
# Create a list for holding the cities.
cities = []
# Identify the nearest city for each latitude and longitude combination.
for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name
    # If the city is unique, then we will add it to the cities list.
    if city not in cities:
        cities.append(city)
# Print the city count to confirm sufficient count.
len(cities)
```

```python
#%%capture   # To hide the output
cities
```

### Perform an API call with the OpenWeatherMap

```python
#1.  Import the requests library. This is to get the file in .Json
import requests
# Import the API key.
from config import weather_api_key
```

```python
# Starting URL for Weather Map API Call.
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key 
print(url)
```

```python
# Import the time library and the datetime module from the datetime library to have a register of the request of API
import time
from datetime import datetime
```

```python
# Second code, using enumerates instead of two for to iterate in i and in city to get the index and the name 
# of the city
# Create an empty list to hold the weather data.
city_data = []
# Print the beginning of the logging.
print("Beginning Data Retrieval     ")
print("-----------------------------")

# Create counters.
record_count = 1
set_count = 1

# step 2
# Loop through all the cities in the list.
# Enumerate allows to iterate through the list of cities and retrieve both the index, and the city from the list.
for i, city in enumerate(cities):

    # Group cities in sets of 50 for logging purposes.
    if (i % 50 == 0 and i >= 50):
        set_count += 1
        record_count = 1
        time.sleep(60)

    # Create endpoint URL with each city.
    city_url = url + "&q=" + city.replace(" ","+")
    #city_url = url + "&q=" + city_data[city]  # Lets see if it works at home

    # Log the URL, record, and set numbers and the city.
    print(f"Processing Record {record_count} of Set {set_count} | {city}")
    # Add 1 to the record count.
    record_count += 1
    
#Step 3
# Run an API request for each of the cities.
    try:
        # Parse the JSON and retrieve data.
        city_weather = requests.get(city_url).json()              # Get all the url
        # Parse out the needed data.
        city_lat = city_weather["coord"]["lat"]                   # We acces to the dictionary of each url to get the
        city_lng = city_weather["coord"]["lon"]                   # required fields
        city_max_temp = city_weather["main"]["temp_max"]
        city_humidity = city_weather["main"]["humidity"]
        city_clouds = city_weather["clouds"]["all"]
        city_wind = city_weather["wind"]["speed"]
        city_country = city_weather["sys"]["country"]
        city_weather_desc = city_weather["weather"][0]["description"]
        # Convert the date to ISO standard.
        city_date = datetime.utcfromtimestamp(city_weather["dt"]).strftime('%Y-%m-%d %H:%M:%S')
        # Append the city information into city_data list.
        city_data.append({"City": city.title(),
                          "Lat": city_lat,
                          "Lng": city_lng,
                          "Max Temp": city_max_temp,
                          "Humidity": city_humidity,
                          "Cloudiness": city_clouds,
                          "Wind Speed": city_wind,
                          "Country": city_country,
                          "Date": city_date,
                          "Current Description": city_weather_desc})

# If an error is experienced, skip the city.
    except:
        print("City not found. Skipping...")
        pass                                                       # Is a genral purpose statement to handle or catch
                                                                   # all errors encountered and allow the program
# Indicate that Data Loading is complete.                          # to continue.
print("-----------------------------")
print("Data Retrieval Complete      ")
print("-----------------------------")
```
